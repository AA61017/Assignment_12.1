```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})

df.info()
df['FlightNumber'] = df['FlightNumber'].interpolate().astype(int)
print("Data frame with corrected FlightNumber")
print(df)

Tempdf = df['From_To'].str.split('_', 1, expand=True).rename(columns={0:'From', 1:'To'})
print("Temp Data frame with the split of From and To")
print(Tempdf)

print("capitalisation of the city names")
Tempdf['From'] = Tempdf['From'].str.capitalize()
Tempdf['To'] = Tempdf['To'].str.capitalize()
print(Tempdf)

print("Joining Temp Data frame with the original")
df = df.drop(['From_To'], axis=1)
df = df.join(Tempdf)
print(df)

print("Expand the Series of lists into a DataFrame named delays")
delay = df['RecentDelays'].apply(pd.Series)
delays = delay.rename(columns = lambda x : 'delay_' + str(x))
print(delays)

print("Replace the unwanted RecentDelays column in d with delays")
df = df.drop(['RecentDelays'], axis=1)
df = df.join(delays)
print(df)
```
